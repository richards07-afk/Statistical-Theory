```python
print("-----8.18-----")
population = [9, 12, 15]
p = 1/3
mu = 0
for x in population:
    mu += x * p
sum_x2 = 0
for x in population:
    sum_x2 += (x**2) * p
var = sum_x2 - mu**2
print("Population mean (μ):", mu)
print("Population variance (σ²):", var)

sample_means = []
for x in population:
    for y in population:
        mean = (x + y) / 2
        sample_means.append(mean)
print("Sample means:", sample_means)

total = 0
for m in sample_means:
    total += m
mu_xbar = total / len(sample_means)

sum_sq = 0
for m in sample_means:
    sum_sq += m**2
var_xbar = (sum_sq / len(sample_means)) - mu_xbar**2
print("Mean of sampling distribution (μ_x̄):", mu_xbar)
print("Variance of sampling distribution (σ²_x̄):", var_xbar)

print("\nRESULTS:")
print("Since μ_x̄ =", mu_xbar, "and μ =", mu, ", they are equal.")
print("Since σ²_x̄ =", var_xbar, "and σ²/2 =", var/2, ", they are equal.\n\n")

print("-----8.21-----")

population = [3,7,11,15]
N = len(population)

total = 0
for x in population:
    total += x
mu = total / N

var_sum = 0
for x in population:
    var_sum += (x - mu)**2
sigma = (var_sum / N) ** 0.5

samples = []
for i in population:
    for j in population:
        samples.append([i,j])

means = []
for s in samples:
    means.append((s[0] + s[1]) / 2)

total_means = 0
for m in means:
    total_means += m
mu_xbar = total_means / len(means)

var_means = 0
for m in means:
    var_means += (m - mu_xbar)**2
sigma_xbar = (var_means / len(means)) ** 0.5

sigma_xbar_formula = sigma / (2 ** 0.5)
print("Population Mean =", mu)
print("Population Standard Deviation =", sigma)
print("\nMean of Sampling Distribution =", mu_xbar)
print("Std Dev of Sampling Distribution =", sigma_xbar)
print("μ = μ_x̄ =", mu_xbar)
print("σ =", sigma, "and n = 2")
print("σ/√n =", sigma_xbar_formula, "which equals", sigma_xbar)
print("\n\n")

print("-----8.34-----")

n = 200
p = 0.5
mean = n*p
sd = (n*p*(1-p))**0.5

def normal_cdf(z):
    t = 1 / (1 + 0.2316419 * abs(z))
    d = 0.3989423 * 2.71828 ** (-z**2 / 2)
    prob = 1 - d * (0.3193815*t - 0.3565638*t**2 + 1.781478*t**3 - 1.821256*t**4 + 1.330274*t**5)
    if z >= 0:
        return prob
    else:
        return 1 - prob

x_a = 80
z_a = (x_a + 0.5 - mean)/sd
prob_a = normal_cdf(z_a)

x_b_low = 86
x_b_high = 114
z_b_low = (x_b_low - 0.5 - mean) / sd
z_b_high = (x_b_high + 0.5 - mean) / sd
prob_b = normal_cdf(z_b_high) - normal_cdf(z_b_low)

x_c = 108
z_c = (x_c - 0.5 - mean) / sd
prob_c = 1 - normal_cdf(z_c)

print("A:", round(prob_a, 4))
print("B:", round(prob_b, 4))
print("C:", round(prob_c, 4))
print("\n\n")

print("------8.49-----")

x = [6, 9, 12, 15, 18]
p = [0.1, 0.2, 0.4, 0.2, 0.1]

mu = 0
for i in range(len(x)):
    mu += x[i] * p[i]

var = 0
for i in range(len(x)):
    var += (x[i] - mu)**2 * p[i]
print("Mean (mu) =", mu)
print("Variance (sigma^2) =", var)
print("\nSamples (size 2 with replacement):")

print(f"{'x1':>4} {'x2':>4} {'mean':>8} {'probability':>12}")
print("-"*32)

for i in range(len(x)):
    for j in range(len(x)):
        x1 = x[i]
        x2 = x[j]
        mean = (x1 + x2) / 2
        prob = p[i] * p[j]
        print(f"{x1:>4} {x2:>4} {mean:>8.1f} {prob:>12.2f}")
```
