```python
import numpy as np
import pandas as pd
from scipy import stats

def solve_assignment():
    print("=== CHI-SQUARE TEST APPLICATIONS: ACCURATE SOLUTIONS ===\n")

    print("## 1. Goodness-of-Fit Test ##")
    obs1 = np.array([24, 108, 95, 23])
    exp1 = np.array([30, 90, 90, 30])

    chi_components1 = ((obs1 - exp1)**2) / exp1
    chi_stat1 = np.sum(chi_components1)
    df1 = len(obs1) - 1
    p_val1 = 1 - stats.chi2.cdf(chi_stat1, df1)

    print(f"a) H0: Coin tosses follow the theoretical distribution.")
    print(f"   Ha: Coin tosses do not follow the theoretical distribution.")
    print(f"b) Chi-square Statistic: {chi_stat1:.2f}")
    print(f"   Degrees of Freedom: {df1}")
    print(f"   P-value: {p_val1:.4f}")
    print(f"c) Decision: {'Reject H0' if p_val1 < 0.05 else 'Fail to Reject H0'}")
    print(f"   Interpretation: The observed frequencies are consistent with the 1:3:3:1 ratio.")

    print("\n## 2. Test of Independence ##")
    data2 = np.array([[9, 42], [17, 28]])
    chi2, p2, dof2, expected2 = stats.chi2_contingency(data2, correction=False)

    print(f"a) Expected Frequencies:\n{expected2}")
    print(f"b) Chi-square Statistic: {chi2:.2f}")
    print(f"   P-value: {p2:.4f}")
    print(f"c) Conclusion: {'Reject H0: Vaccination is related to disease.' if p2 < 0.05 else 'Fail to Reject: Vaccination and disease are independent.'}")

    print("\n## 3. Real-World Context ##")
    data3 = np.array([[44, 10], [81, 35]])
    chi3, p3, dof3, expected3 = stats.chi2_contingency(data3, correction=False)

    print(f"a) Chi-square Statistic: {chi3:.2f}")
    print(f"   P-value: {p3:.4f}")
    print(f"b) Interpretation: {'The pills have a significant effect.' if p3 < 0.05 else 'No significant evidence that pills are more effective than sugar pills.'}")

    print("\n## 4. Measure of Association ##")
    data4 = np.array([[49, 25], [30, 96]])
    chi4, p4, dof4, expected4 = stats.chi2_contingency(data4, correction=False)

    n = np.sum(data4)
    c_val = np.sqrt(chi4 / (chi4 + n))

    print(f"a) Coefficient of Contingency (C): {c_val:.3f}")
    print(f"   P-value (from Chi-Square): {p4:.4f}")

    strength = "Weak" if c_val < 0.3 else "Moderate" if c_val < 0.5 else "Strong"
    print(f"b) Strength: This represents a {strength} association.")

solve_assignment()
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

def solve_curve_fitting_assignment():
    print("=== CURVE FITTING AND LEAST SQUARES: ACCURATE SOLUTIONS ===\n")

    print("## 1. Straight Lines ##")
    x1_pts, y1_pts = np.array([3, -1]), np.array([-2, 6])
    m1, b1 = np.polyfit(x1_pts, y1_pts, 1)

    print(f"a) Slope (m): {m1:.2f}")
    print(f"b) Equation: y = {m1:.2f}x + {b1:.2f}")
    print(f"c) X-intercept: {-b1/m1:.2f}, Y-intercept: {b1:.2f}")
    print(f"d) Verification: Point 1: {m1*3+b1:.2f}, Point 2: {m1*-1+b1:.2f}")

    print("\n## 2. Least Squares Line ##")
    x2 = np.array([3, 5, 6, 8, 9, 11])
    y2 = np.array([2, 3, 4, 6, 5, 8])

    slope2, intercept2, r_val2, p_val2, std_err2 = stats.linregress(x2, y2)

    print(f"a) ΣX: {sum(x2)}, ΣY: {sum(y2)}, ΣXY: {sum(x2*y2)}, ΣX²: {sum(x2**2)}")
    print(f"b) Regression Equation: Y = {intercept2:.2f} + {slope2:.2f}X")
    print(f"c) Prediction for X=5: {intercept2 + slope2*5:.2f}")
    print(f"   Prediction for X=12: {intercept2 + slope2*12:.2f}")
    print(f"d) Slope Interpretation: For every unit increase in X, Y increases by {slope2:.2f}")
    print(f"   P-value (Significance of slope): {p_val2:.4f}")

    print("\n## 3. Real-World Application ##")
    alg_x = np.array([75, 80, 93, 65, 87, 71, 98, 68, 84, 77])
    phy_y = np.array([82, 78, 86, 72, 91, 80, 95, 72, 89, 74])

    slope3, intercept3, r_val3, p_val3, std_err3 = stats.linregress(alg_x, phy_y)

    print(f"b) Equation: Physics = {intercept3:.2f} + {slope3:.2f}(Algebra)")
    print(f"c) Predicted Physics grade (Algebra=75): {intercept3 + slope3*75:.2f}")
    print(f"d) Predicted Algebra grade (Physics=95): {(95 - intercept3) / slope3:.2f}")
    print(f"e) Correlation (r): {r_val3:.4f}")
    print(f"   P-value: {p_val3:.4f}")

    print("\n## 4. Least Squares Curve ##")
    x4 = np.array([0, 1, 2, 3, 4, 5, 6])
    y4 = np.array([2.4, 2.1, 3.2, 5.6, 9.3, 14.6, 21.9])

    a2, a1, a0 = np.polyfit(x4, y4, 2)

    print(f"a) Quadratic Model: Y = {a0:.2f} + {a1:.2f}X + {a2:.2f}X²")
    print(f"b) Predicted Y for X=7: {a0 + a1*7 + a2*(7**2):.2f}")
    print(f"c) Trend: The data shows an increasing rate of change, making the quadratic fit appropriate.")

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.scatter(alg_x, phy_y, color='blue')
    plt.plot(alg_x, intercept3 + slope3*alg_x, color='red')
    plt.title("Algebra vs Physics Regression")

    plt.subplot(1, 2, 2)
    plt.scatter(x4, y4, color='green')
    x_range = np.linspace(0, 7, 100)
    plt.plot(x_range, a0 + a1*x_range + a2*x_range**2, color='orange')
    plt.title("Quadratic Curve Fit")
    plt.show()

solve_curve_fitting_assignment()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

print("=== MULTIPLE AND PARTIAL CORRELATION ===\n\n")

def solve_full_assignment():
    data1 = {
        'X1': [3, 5, 6, 8, 12, 14],
        'X2': [16, 10, 7, 4, 3, 2],
        'X3': [90, 72, 54, 42, 30, 12]
    }
    df1 = pd.DataFrame(data1)

    X12 = sm.add_constant(df1[['X1', 'X2']])
    y3 = df1['X3']
    model1 = sm.OLS(y3, X12).fit()

    a, b1, b2 = model1.params
    print("--- 1. Multiple Regression Equation ---")
    print(f"a. Equation: X3 = {a:.2f} + ({b1:.2f})X1 + ({b2:.2f})X2")

    est_x3 = model1.predict([1, 10, 6])[0]
    print(f"b. Estimated X3 when X1=10 and X2=6: {est_x3:.2f}\n")

    se_est = np.sqrt(model1.mse_resid)
    print("--- 2. Standard Error of Estimate ---")
    print(f"a. Standard error of estimate: {se_est:.2f}\n")

    r_mult = np.sqrt(model1.rsquared)
    print("--- 3. Coefficient of Multiple Correlation ---")
    print(f"a. R value: {r_mult:.2f}\n")

    corrs = df1.corr()
    r12, r13, r23 = corrs.loc['X1','X2'], corrs.loc['X1','X3'], corrs.loc['X2','X3']
    r12_3 = (r12 - (r13 * r23)) / np.sqrt((1 - r13**2) * (1 - r23**2))
    print("--- 4. Partial Correlation ---")
    print(f"a. Partial correlation r12.3: {r12_3:.2f}\n")

    housing = {
        'Price': [165, 200, 225, 180, 202, 250, 275, 300, 155, 230],
        'Bedrooms': [3, 3, 4, 2, 4, 4, 3, 5, 2, 4],
        'Baths': [2, 3, 3, 3, 2, 4, 4, 3, 2, 4]
    }
    df_h = pd.DataFrame(housing)
    X_h = sm.add_constant(df_h[['Bedrooms', 'Baths']])
    y_h = df_h['Price']
    model_h = sm.OLS(y_h, X_h).fit()

    ha, hb1, hb2 = model_h.params
    print("--- 5. Real-World Multiple Regression ---")
    print(f"a. Model: Price = {ha:.2f} + ({hb1:.2f})Bedrooms + ({hb2:.2f})Baths")

    est_price = model_h.predict([1, 5, 4])[0]
    print(f"b. Estimated Price (5 Bed, 4 Bath): {est_price:.2f}\n")

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    axes[0, 0].scatter(model1.predict(X12), y3, color='blue', edgecolor='k')
    axes[0, 0].plot([y3.min(), y3.max()], [y3.min(), y3.max()], 'r--')
    axes[0, 0].set_title('Problem 1: Actual vs Predicted X3')
    axes[0, 0].set_xlabel('Predicted'); axes[0, 0].set_ylabel('Actual')

    axes[0, 1].scatter(model1.predict(X12), model1.resid, color='purple', edgecolor='k')
    axes[0, 1].axhline(y=0, color='black', linestyle='-')
    axes[0, 1].set_title('Problem 2: Residual Plot (Error Analysis)')
    axes[0, 1].set_xlabel('Predicted'); axes[0, 1].set_ylabel('Residuals')

    sns.heatmap(df1.corr(), annot=True, cmap='RdYlGn', ax=axes[1, 0])
    axes[1, 0].set_title('Problem 4: Dataset 1 Correlation Heatmap')

    axes[1, 1].scatter(model_h.predict(X_h), y_h, color='green', edgecolor='k')
    axes[1, 1].plot([y_h.min(), y_h.max()], [y_h.min(), y_h.max()], 'r--')
    axes[1, 1].set_title('Problem 5: Housing Price Fit')
    axes[1, 1].set_xlabel('Predicted Price'); axes[1, 1].set_ylabel('Actual Price')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    solve_full_assignment()
```
